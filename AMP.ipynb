```python
import math as mt
import random as rm
import matplotlib.pyplot as plt 
x0=rm.uniform(-15,15)
x=x0
y=4.9
t=0
g=-9.78
r=0.35
dt=0.001
x1=0
y1=0
xl=[]
yl=[]
```

```python
def csn(x,y,r) :
    if mt.sqrt(((x-round(x))**2)+(y-round(y))**2)<=2*r :
        if (round(x)%2==0 and round(y)%2==0) or (round(x)%2==1 and round(y)%2==1) :
            o=1
        else :
            o=0
    else:
        o=0
    if y<=5 and y>=0 :
        o=0

    return(o)
```

```python
def aht(x,y,x1,y1) :
    cs=(x-round(x))/(mt.sqrt((x-round(x))**2+(y-round(y))**2))
    sn=(y-round(y))/(mt.sqrt((x-round(x))**2+(y-round(y))**2))
    x1p=-((x1*cs)+(y1*sn))
    y1p=(-x1*sn)+(y1*cs)
    x1f=(x1p*cs)-(y1p*sn)
    y1f=(x1p*sn)+(y1p*cs)
    y=y+(y1f*0.001)
    x=x+(x1f*0.001)
    return(x,y,x1f,y1f)
```

```python
def pjn(x,y,t,x1,y1,xl,yl) :
    while y<=5 and y>=-100 and csn(x,y,r)==0 :
        x=x+(x1*dt)
        y1=y1+(g*dt)
        y=y+(y1*dt)
        xl.append(x)
        yl.append(y)
        t+=dt
        print(csn(x,y,r),x,y,x1,y1)
    return(x,y,x1,y1,xl,yl,t)
```

```python

while y>=-100 :
    if y==5:
        y1=-y1
    (x,y,x1,y1,xl,yl,t)=pjn(x,y,t,x1,y1,xl,yl)
    (x,y,x1,y1)=aht(x,y,x1,y1)
plt.plot(xl, yl)
plt.xlabel('x - axis')
plt.ylabel('y - axis') 
plt.title('Projection')
plt.show()
```
