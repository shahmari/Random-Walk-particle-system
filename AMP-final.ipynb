```python
import math as mt
import random as rm
import matplotlib.pyplot as plt 
y=4.9
g=-5
r=0.1
dt=0.01
x1=0
y1=0
xl=[]
yl=[]
```

```python
def csn(x,y,r) :
    xr=round(x)
    yr=round(y)
    if mt.sqrt((x-xr)**2+(y-yr)**2)<=2*r :
        if (xr%2==0 and yr%2==0) or (xr%2==1 and yr%2==1) :
            o=1
        else :
            o=0
    else:
        o=0
    if y>2*r :
        o=0
    return(o)
```
